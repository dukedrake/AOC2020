# Part 1

```python
def part1():
    otherField = []
    fieldCopy = []
    for cRow in field:
        cRow = [int(x) for x in cRow]
        fieldCopy.append(cRow)
    
    epoch = 1
    while True:
        otherField = calcEpoch(fieldCopy)
        
        if checkFieldDiff(fieldCopy,otherField):
            numSeats = 0
            for row in otherField:
                for cell in row:
                    numSeats += 1 if cell == 2 else 0
            # print('done','# of seats:',numSeats,'epoch:',epoch)
            # printField(otherField) 
            break
        
        else:
            fieldCopy = otherField
        epoch += 1
        
        if epoch > 99:
            break

def printField(field):
    print('-------------------')
    for row in field:
        for cell in row:
            print(cell, end=' ')
        print('')

        
# only down to row level
# return False if fields are not identical
def checkFieldDiff(fieldA,fieldB):
    checkRowCounter = 0
    while checkRowCounter < len(fieldA):
        if sum(fieldA[checkRowCounter]) != sum(fieldB[checkRowCounter]):
            return False
        checkRowCounter += 1
    return True

        
    
def calcEpoch(currentField):
    copyField = []
    for row in currentField:
        copyField.append(row.copy())
    rowCounter = 0
    while rowCounter < len(currentField):
        minY = rowCounter - 1 if rowCounter > 0 else rowCounter
        maxY = rowCounter + 1 if rowCounter < len(currentField) -1 else rowCounter
        
        cellCounter = 0
        while cellCounter < len(currentField[rowCounter]):
            curCell = currentField[rowCounter][cellCounter]
            
            # if current cell is no floor
            if curCell:
                minX = cellCounter - 1 if cellCounter > 0 else cellCounter
                maxX = cellCounter + 1 if cellCounter < len(currentField[rowCounter]) -1 else cellCounter
                
                # test adjacent cells
                numOccupied = 0
                for y in range(minY,maxY+1):
                    for x in range(minX,maxX+1): 
                        if y == rowCounter and x == cellCounter:
                            continue
                        if currentField[y][x] == 2:
                            numOccupied += 1
                
                if numOccupied >= 4:
                    copyField[rowCounter][cellCounter] = 1
                if numOccupied == 0:
                    copyField[rowCounter][cellCounter] = 2
                
            cellCounter += 1
        rowCounter += 1
        
    return copyField

        
%time part1()

# n = 100
# %time for i in range(n): part1()
```

```python
def part2():
    otherField = []
    fieldCopy = []
    for cRow in field:
        cRow = [int(x) for x in cRow]
        fieldCopy.append(cRow)
    
    epoch = 1
    while True:
        otherField = calcEpochPart2(fieldCopy)
        
        if checkFieldDiff(fieldCopy,otherField):
            numSeats = 0
            for row in otherField:
                for cell in row:
                    numSeats += 1 if cell == 2 else 0
            print('done','# of seats:',numSeats,'epoch:',epoch)
            # printField(otherField) 
            break
        
        else:
            fieldCopy = otherField
        epoch += 1
        
        if epoch > 200:
            print('epoch maxed out')
            break

            
def printField(field):
    print('-------------------')
    for row in field:
        for cell in row:
            if not cell:
                print('.', end=' ')
            if cell == 2:
                print('#', end=' ')
            if cell == 1:
                print('L', end=' ')
        print('')

        
# only down to row level
# return False if fields are not identical
def checkFieldDiff(fieldA,fieldB):
    checkRowCounter = 0
    while checkRowCounter < len(fieldA):
        if sum(fieldA[checkRowCounter]) != sum(fieldB[checkRowCounter]):
            return False
        checkRowCounter += 1
    return True

            
def calcEpochPart2(currentField):
    copyField = []
    for row in currentField:
        copyField.append(row.copy())
    rowCounter = 0
    while rowCounter < len(currentField):
        upRelYOff = -1 if rowCounter > 0 else 0
        downRelYOff = 1 if rowCounter < len(currentField) -1 else 0
        
        cellCounter = 0
        while cellCounter < len(currentField[rowCounter]):
            curCell = currentField[rowCounter][cellCounter]
            # if current cell is no floor
            if curCell:
                leftRelXOff = -1 if cellCounter > 0 else 0
                rightRelXOff = 1 if cellCounter < len(currentField[rowCounter]) -1 else 0
                
                # test adjacent cells
                numOccupied = 0
                for y in range(upRelYOff,downRelYOff+1):
                    for x in range(leftRelXOff,rightRelXOff+1): 
                        
                        absX = x + cellCounter
                        tmpX = absX
                        absY = y + rowCounter
                        tmpY = absY
                        
                        if absY == rowCounter and absX == cellCounter:
                            continue
                            
                        # offset in directions array
                        arrOff = x+1 + (y+1) * 3
                        
                        while tmpX >=0 and tmpX < len(currentField[rowCounter]) and tmpY >=0 and tmpY < len(currentField):
                            if currentField[tmpY][tmpX]:
                                if currentField[tmpY][tmpX] == 2:
                                    numOccupied += 1
                                break
                            # next Test
                            tmpX += x
                            tmpY += y
                        
                if numOccupied >= 5:
                    copyField[rowCounter][cellCounter] = 1
                if numOccupied == 0:
                    copyField[rowCounter][cellCounter] = 2
                
            cellCounter += 1
        rowCounter += 1
        
    return copyField


%time part2()
```

```python
def print_factors(x):
    print("The factors of",x,"are:")
    for i in range(1, x + 1):
        if x % i == 0:
            print(i)

def primes(n):
    primfac = []
    d = 2
    while d*d <= n:
        while (n % d) == 0:
            primfac.append(d)  # supposing you want multiple factors repeated
            n //= d
        d += 1
    if n > 1:
       primfac.append(n)
    return primfac
```
